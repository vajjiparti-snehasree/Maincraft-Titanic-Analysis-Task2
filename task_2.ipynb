```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('titanic')

df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['age'].fillna(df['age'].median(), inplace=True)
```

```python
df['age'] = df['age'].fillna(df['age'].median())
```

```python
df['age'].isnull().sum()
```

```python
df.groupby('sex')['survived'].mean()
```

```python
df.groupby('sex')['survived'].mean()
```

# Survival Analysis by Gender
Calculated the survival rate of male and female passengers using groupby() and mean().

Female passengers had a higher survival rate compared to male passengers.

```python
df.groupby('class')['survived'].mean()
```

# Survival Analysis by Passenger Class
Analyzed survival rates based on passenger class.

Passengers in first class had higher survival chances compared to lower classes.

```python
df['age_group'] = pd.cut(df['age'],
                         bins=[0,12,19,35,60,100],
                         labels=['Child','Teen','Young Adult','Adult','Senior'])
```

```python
df.groupby('age_group')['survived'].mean()
```

# Survival Rate by Age Group
Created age categories and analyzed survival rates across different age groups.

```python
sns.barplot(x='sex', y='survived', data=df)

plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")

plt.show()
```

```python
sns.barplot(x='class', y='survived', data=df)

plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")

plt.show()
```

# Bar Chart: Survival by Passenger Class
Compared survival rates across different passenger classes.

```python
plt.hist(df['age'])

plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")

plt.show()
```
